```python
import openpyxl
import requests

wb = openpyxl.load_workbook('./data/GraceDB.xlsx')
ws = wb['Sheet1']
# This will fail if there is no hyperlink to target
nrow = ws.max_row
ncol = ws.max_column

FlatName = []

for i in range(nrow-1):
    cell = ws.cell(row=i+2,column=1)
    eventID = cell.value
    if ws.cell(row=i+2,column=5).hyperlink is not None:
        url = ws.cell(row=i+2,column=5).hyperlink.target
#         print(url)
        f = requests.get(url)
         #获取文件类型并确定文件名
        smName = ws.cell(row=i+2,column=5).value
        suffix = smName[smName.find('.'):]
        fileName = r"./data/SkyMap/Flat/"+eventID+"_Flat"+suffix
        FlatName.append(eventID+"_Flat"+suffix)
        print(eventID+"_Flat"+suffix)
        with open(fileName, "wb") as code:
            code.write(f.content)
    else:
        print('No hyperlink found in the cell')
    
```
